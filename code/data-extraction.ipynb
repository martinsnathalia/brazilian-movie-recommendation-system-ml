```python
#def raspa_filme(start,stop):

ano = []
titulo = []
diretor = []
watchedby = []
listas = []
liked = []
ranking = []
fans = []
rating = []
duracao = []
sinopse = []
elenco = []
genero = []
temas = []
pais = []
titulo_eua = []
titulo_br = []
tag_tema_1 = []
tag_tema_2 = []
tag_tema_3 = []
tag_tema_4 = []
tag_tema_5 = []
genero1 = []
genero2 = []
genero3 = []

    #filmes_links = pagina_filme[start:stop]
for i in paginas[:5]:  
    
    wd_chrome.get(f"https://letterboxd.com{i}")
    pageSource_filme =wd_chrome.page_source
    bs = BeautifulSoup(pageSource_filme, 'html.parser')
    sleep(5)

    for item in bs: # para cada um dos elementos retornados:

        ano.append(item.find_all('small', class_="number")[0].get_text())
        titulo.append(item.findChildren('section', class_="film-header-lockup -default")[0].get_text().strip().split("\n"))
        diretor.append(item.find_all('span', class_="prettify")[0].get_text())
        watchedby.append(item.find_all('li', class_="stat filmstat-watches")[0].get_text())
        listas.append(item.find_all('li', class_="stat filmstat-lists")[0].get_text())
        liked.append(item.find_all('li', class_="stat filmstat-likes")[0].get_text())
        try:
            fans.append(item.find_all('a', class_="all-link more-link")[0].get_text().split()[0])
        except:
            fans.append('0')
        duracao.append(item.find_all('p', class_="text-link text-footer")[0].get_text().strip().split()[0])
        try:
            sinopse.append(item.find_all('div', class_="truncate")[0].get_text().strip())
        except:
            sinopse.append(' ')
        #elenco.append(item.findChildren('div', class_="cast-list text-sluglist")[0].get_text().strip())

        loop_elenco = len(item.find_all('a', href=lambda href: href and href.startswith("/actor/")))
        cast = []
        for i in range(0, loop_elenco):
            cast.append(item.find_all('a', href=lambda href: href and href.startswith("/actor/"))[i].get_text().strip())
        elenco.append(cast) 
        
        status_ranking = item.find_all('li', class_="stat filmstat-top250")
        if len(status_ranking) > 0:
            ranking.append(item.find_all('li', class_="stat filmstat-top250")[0].get_text().strip())
        else: 
            ranking.append(' ')
            
        #rate = item.find_all('a', class_="tooltip display-rating")
        if len(item.find_all('a', class_="tooltip display-rating")) > 0:
            rating.append(item.find_all('a', class_="tooltip display-rating")[0].get_text())
        elif len(item.find_all('a', class_="tooltip display-rating -highlight")) > 0:
            rating.append(item.find_all('a', class_="tooltip display-rating -highlight")[0].get_text())
        else: 
            rating.append(' ')
        
        # País
        loop_pais = len(item.find_all('a', href=lambda href: href and href.startswith("/films/country/")))
        country = []
        for i in range(0, loop_pais):
            country.append(item.find_all('a', href=lambda href: href and href.startswith("/films/country/"))[i].get_text().strip())
        pais.append(country)
        
        #Gênero
        gen = []
        loop_genero = len(item.find_all('a', href=lambda href: href and href.startswith("/films/genre/")))
        for i in range(0, loop_genero):
            gen.append(item.find_all('a', href=lambda href: href and href.startswith("/films/genre/"))[i].get_text().strip())
        genero.append(gen)
      
        # Temas
        
        loop_temas_1 = len(item.find_all('a', href=lambda href: href and href.startswith("/films/theme/")))
        loop_temas_2 = len(item.find_all('a', href=lambda href: href and href.startswith("/films/mini-theme/")))
        temas_tags = []
        temas_status = item.find_all('a', href=lambda href: href and (href.startswith("/films/theme/")))
        if len(temas_status) > 0:
            for i in range(0, loop_temas_1):
                temas_tags.append(item.find_all('a', href=lambda href: href and (href.startswith("/films/theme/")))[i].get_text().strip())

            for i in range(0,loop_temas_2):
                temas_tags.append(item.find_all('a', href=lambda href: href and (href.startswith("/films/mini-theme/")))[i].get_text().strip())
            temas.append(temas_tags)
        else:
            temas.append('--')
        
    #return ano, titulo, diretor, watchedby, listas,liked, ranking, fans, rating, duracao, sinopse, elenco, genero, temas, pais
```
